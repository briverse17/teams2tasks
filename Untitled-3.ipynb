# Task Kanban Board Visualization

This notebook loads inferred tasks from Teams chat conversations and visualizes them as an interactive Kanban board with filtering capabilities.

**Features:**
- Interactive Kanban board organized by status (Pending, In Progress, Completed)
- Visual indicators for priority levels and assignees
- Dynamic filtering by assignee, priority, chat channel, and due date range
- Task detail exploration and statistics

```python
# Section 1: Import Required Libraries
import json
from datetime import datetime

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots

# Set global dark theme for all Plotly charts
pio.templates.default = "plotly_dark"

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

```python
# Section 2: Load Task Data
# Load the tasks_inferred_specific.json file
json_file_path = r"c:\workspace\scrape-teams-chats\outputs\tasks_inferred_specific.json"

with open(json_file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data["tasks"])

print(f"Total tasks loaded: {len(df)}")
print(f"\nDataFrame shape: {df.shape}")
print(f"\nColumn names: {df.columns.tolist()}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Section 3: Prepare Data for Visualization

# Handle missing values and standardize data types
df["due_date"] = pd.to_datetime(df["due_date"], errors="coerce")
df["status"] = df["status"].str.strip().str.title()

# Standardize status values to match Kanban columns
status_mapping = {
    "Pending": "Pending",
    "In Progress": "In Progress",
    "In progress": "In Progress",
    "Completed": "Completed",
    "Todo": "Pending",
}
df["status"] = df["status"].map(status_mapping).fillna(df["status"])

# Handle missing priority
df["priority"] = df["priority"].str.strip().str.title()
priority_mapping = {"High": "High", "Medium": "Medium", "Low": "Low"}
df["priority"] = df["priority"].map(priority_mapping).fillna("Medium")

# Calculate days until due date
today = datetime.now()
df["days_until_due"] = (df["due_date"] - pd.Timestamp(today)).dt.days
df["is_overdue"] = df["days_until_due"] < 0
df["due_date_display"] = df["due_date"].dt.strftime("%Y-%m-%d").fillna("No due date")

# Create priority numeric values for sorting
priority_order = {"High": 1, "Medium": 2, "Low": 3}
df["priority_order"] = df["priority"].map(priority_order)

# Status order for Kanban columns
status_order = {"Pending": 1, "In Progress": 2, "Completed": 3}
df["status_order"] = df["status"].map(status_order)

print("Data preparation complete!")
print(f"\nStatus distribution:")
print(df["status"].value_counts())
print(f"\nPriority distribution:")
print(df["priority"].value_counts())
print(f"\nChat channels:")
print(df["chat_name"].value_counts())
```

```python
# Section 4: Create Kanban Board Visualization

# Define color schemes
status_colors = {
    "Pending": "#FFA500",  # Orange
    "In Progress": "#4169E1",  # Royal Blue
    "Completed": "#28A745",  # Green
}

priority_colors = {
    "High": "#DC143C",  # Crimson
    "Medium": "#FF6347",  # Tomato
    "Low": "#90EE90",  # Light Green
}

# Prepare data for Kanban board
kanban_data = df.sort_values(["status_order", "priority_order", "due_date"])

# Create status columns
statuses = ["Pending", "In Progress", "Completed"]

# Create figure with subplots
fig = go.Figure()

# Add traces for each status column
x_positions = {status: i for i, status in enumerate(statuses)}

for status in statuses:
    status_df = kanban_data[kanban_data["status"] == status]

    # Create hover text with task details
    hover_texts = []
    for idx, row in status_df.iterrows():
        hover_text = (
            f"<b>{row['title']}</b><br>"
            f"ID: {row['id']}<br>"
            f"Assignee: {row['assignee']}<br>"
            f"Priority: {row['priority']}<br>"
            f"Due: {row['due_date_display']}<br>"
            f"Chat: {row['chat_name']}<br>"
            f"Description: {row['description'][:100]}..."
        )
        hover_texts.append(hover_text)

    # Create scatter plot for this status
    fig.add_trace(
        go.Scatter(
            x=[x_positions[status]] * len(status_df),
            y=list(range(len(status_df))),
            mode="markers+text",
            marker=dict(
                size=25,
                color=[priority_colors[p] for p in status_df["priority"]],
                opacity=0.7,
                line=dict(width=2, color="white"),
            ),
            text=status_df["title"].str[:20],
            textposition="middle center",
            textfont=dict(size=9, color="white", family="Arial Black"),
            hovertext=hover_texts,
            hoverinfo="text",
            name=status,
            showlegend=False,
        )
    )

# Update layout with dark theme
fig.update_layout(
    title={
        "text": "<b>Task Kanban Board</b><br><sub>Interactive visualization of inferred tasks by status</sub>",
        "x": 0.5,
        "xanchor": "center",
        "font": {"size": 24, "color": "#ffffff"},
    },
    xaxis=dict(
        tickvals=list(x_positions.values()),
        ticktext=list(x_positions.keys()),
        showgrid=False,
        zeroline=False,
        tickfont=dict(size=14, family="Arial Black", color="#ffffff"),
        side="top",
    ),
    yaxis=dict(showticklabels=False, showgrid=False, zeroline=False),
    plot_bgcolor="#1e1e1e",
    paper_bgcolor="#111111",
    height=600,
    width=1200,
    hovermode="closest",
    margin=dict(l=50, r=50, t=100, b=50),
)

fig.show()

print(f"\nKanban Board Statistics:")
print(f"Total tasks: {len(kanban_data)}")
for status in statuses:
    count = len(kanban_data[kanban_data["status"] == status])
    print(f"{status}: {count} tasks")
```

```python
# Section 5: Add Interactivity and Filtering

# Create interactive dashboard with filtering

# Get unique values for filters
assignees = sorted(df["assignee"].unique())
priorities = ["High", "Medium", "Low"]
chat_channels = sorted(df["chat_name"].unique())

print("Available filters:")
print(f"Assignees: {assignees}")
print(f"Priorities: {priorities}")
print(f"Chat channels: {chat_channels}")
```

```python
# Create a function to generate filtered Kanban board
def create_filtered_kanban(
    dataframe,
    selected_assignees=None,
    selected_priorities=None,
    selected_chats=None,
    selected_statuses=None,
    show_overdue_only=False,
):
    """
    Create an interactive Kanban board with applied filters.

    Parameters:
    - dataframe: The task dataframe
    - selected_assignees: List of assignees to include (None = all)
    - selected_priorities: List of priorities to include (None = all)
    - selected_chats: List of chat channels to include (None = all)
    - selected_statuses: List of statuses to include (None = all)
    - show_overdue_only: Boolean to show only overdue tasks
    """

    filtered_df = dataframe.copy()

    # Apply filters
    if selected_assignees:
        filtered_df = filtered_df[filtered_df["assignee"].isin(selected_assignees)]

    if selected_priorities:
        filtered_df = filtered_df[filtered_df["priority"].isin(selected_priorities)]

    if selected_chats:
        filtered_df = filtered_df[filtered_df["chat_name"].isin(selected_chats)]

    if selected_statuses:
        filtered_df = filtered_df[filtered_df["status"].isin(selected_statuses)]

    if show_overdue_only:
        filtered_df = filtered_df[filtered_df["is_overdue"] == True]

    # Sort for Kanban
    filtered_df = filtered_df.sort_values(
        ["status_order", "priority_order", "due_date"]
    )

    # Create figure
    fig = go.Figure()
    statuses = ["Pending", "In Progress", "Completed"]
    x_positions = {status: i for i, status in enumerate(statuses)}

    for status in statuses:
        status_df = filtered_df[filtered_df["status"] == status]

        if len(status_df) == 0:
            continue

        # Create hover text
        hover_texts = []
        for idx, row in status_df.iterrows():
            overdue_badge = "🚨 OVERDUE" if row["is_overdue"] else ""
            hover_text = (
                f"<b>{row['title']}</b> {overdue_badge}<br>"
                f"ID: {row['id']}<br>"
                f"Assignee: {row['assignee']}<br>"
                f"Priority: <b>{row['priority']}</b><br>"
                f"Due: {row['due_date_display']}"
                f"{' (OVERDUE)' if row['is_overdue'] else ''}<br>"
                f"Chat: {row['chat_name']}<br>"
                f"Status: {row['status']}<br>"
                f"Description: {row['description'][:80]}..."
            )
            hover_texts.append(hover_text)

        fig.add_trace(
            go.Scatter(
                x=[x_positions[status]] * len(status_df),
                y=list(range(len(status_df))),
                mode="markers+text",
                marker=dict(
                    size=28,
                    color=[priority_colors[p] for p in status_df["priority"]],
                    opacity=0.8,
                    line=dict(
                        width=2,
                        color=(
                            "#333333" if not any(status_df["is_overdue"]) else "#DC143C"
                        ),
                    ),
                    symbol=[
                        "diamond" if row["is_overdue"] else "circle"
                        for _, row in status_df.iterrows()
                    ],
                ),
                text=status_df["title"].str[:18],
                textposition="middle center",
                textfont=dict(size=8, color="white", family="Arial Black"),
                hovertext=hover_texts,
                hoverinfo="text",
                name=status,
                showlegend=False,
            )
        )

    fig.update_layout(
        title={
            "text": f"<b>Filtered Task Kanban Board</b><br><sub>Showing {len(filtered_df)} of {len(dataframe)} tasks</sub>",
            "x": 0.5,
            "xanchor": "center",
            "font": {"size": 20, "color": "#ffffff"},
        },
        xaxis=dict(
            tickvals=list(x_positions.values()),
            ticktext=list(x_positions.keys()),
            showgrid=False,
            zeroline=False,
            tickfont=dict(size=12, family="Arial Black", color="#ffffff"),
            side="top",
        ),
        yaxis=dict(showticklabels=False, showgrid=False, zeroline=False),
        plot_bgcolor="#1e1e1e",
        paper_bgcolor="#111111",
        height=600,
        width=1200,
        hovermode="closest",
        margin=dict(l=50, r=50, t=120, b=50),
    )

    return fig, filtered_df


# Example 1: Show only High priority tasks
print("\n=== Filter Example 1: High Priority Tasks ===")
fig1, filtered_df1 = create_filtered_kanban(df, selected_priorities=["High"])
fig1.show()
print(f"Showing {len(filtered_df1)} high priority tasks")
```

```python
# Example 2: Show tasks by specific assignee
print("\n=== Filter Example 2: Tasks assigned to Nguyen Minh Vu ===")
fig2, filtered_df2 = create_filtered_kanban(df, selected_assignees=["Nguyen Minh Vu"])
fig2.show()
print(f"Showing {len(filtered_df2)} tasks assigned to Nguyen Minh Vu")
```

```python
# Example 3: Show In Progress and Pending tasks only
print("\n=== Filter Example 3: Active Tasks (In Progress + Pending) ===")
fig3, filtered_df3 = create_filtered_kanban(
    df, selected_statuses=["In Progress", "Pending"]
)
fig3.show()
print(f"Showing {len(filtered_df3)} active tasks")
```

```python
# Example 4: Show overdue tasks
print("\n=== Filter Example 4: Overdue Tasks ===")
overdue_tasks = df[df["is_overdue"] == True]
fig4, filtered_df4 = create_filtered_kanban(df, show_overdue_only=True)
fig4.show()
print(f"Total overdue tasks: {len(overdue_tasks)}")
print(overdue_tasks[["title", "due_date_display", "status", "priority", "assignee"]])
```

```python
# Create summary statistics dashboard
fig_stats = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=(
        "Tasks by Status",
        "Tasks by Priority",
        "Tasks by Assignee",
        "Tasks by Chat Channel",
    ),
    specs=[[{"type": "pie"}, {"type": "pie"}], [{"type": "bar"}, {"type": "bar"}]],
)

# Status pie chart
status_counts = df["status"].value_counts()
fig_stats.add_trace(
    go.Pie(labels=status_counts.index, values=status_counts.values, name="Status"),
    row=1,
    col=1,
)

# Priority pie chart
priority_counts = (
    df["priority"].value_counts().reindex(["High", "Medium", "Low"], fill_value=0)
)
fig_stats.add_trace(
    go.Pie(
        labels=priority_counts.index,
        values=priority_counts.values,
        marker=dict(colors=["#DC143C", "#FF6347", "#90EE90"]),
        name="Priority",
    ),
    row=1,
    col=2,
)

# Assignee bar chart
assignee_counts = df["assignee"].value_counts().head(10)
fig_stats.add_trace(
    go.Bar(
        y=assignee_counts.index,
        x=assignee_counts.values,
        orientation="h",
        marker=dict(color="#4169E1"),
        showlegend=False,
    ),
    row=2,
    col=1,
)

# Chat channel bar chart
chat_counts = df["chat_name"].value_counts().head(10)
fig_stats.add_trace(
    go.Bar(
        y=chat_counts.index,
        x=chat_counts.values,
        orientation="h",
        marker=dict(color="#FF8C00"),
        showlegend=False,
    ),
    row=2,
    col=2,
)

fig_stats.update_layout(
    title_text="<b>Task Statistics Dashboard</b>",
    title_font_color="#ffffff",
    showlegend=False,
    height=700,
    width=1200,
    plot_bgcolor="#1e1e1e",
    paper_bgcolor="#111111",
)
fig_stats.update_xaxes(title_text="Count", title_font_color="#ffffff", tickfont_color="#ffffff", row=2, col=1)
fig_stats.update_xaxes(title_text="Count", title_font_color="#ffffff", tickfont_color="#ffffff", row=2, col=2)
fig_stats.update_yaxes(title_font_color="#ffffff", tickfont_color="#ffffff")

fig_stats.show()
```

```python
# Create a detailed task report by status
print("\n" + "=" * 100)
print("DETAILED TASK REPORT BY STATUS")
print("=" * 100)

for status in ["Pending", "In Progress", "Completed"]:
    status_tasks = df[df["status"] == status].sort_values("priority_order")
    print(f"\n{status.upper()} ({len(status_tasks)} tasks)")
    print("-" * 100)

    for idx, (_, row) in enumerate(status_tasks.iterrows(), 1):
        overdue_badge = " 🚨 OVERDUE" if row["is_overdue"] else ""
        print(f"{idx}. [{row['priority']}] {row['title']}{overdue_badge}")
        print(f"   ID: {row['id']} | Assignee: {row['assignee']}")
        print(f"   Due: {row['due_date_display']} | Chat: {row['chat_name']}")
        print(f"   Description: {row['description']}")
        print()
```

```python
# Create timeline visualization - tasks by due date
df_with_due_date = df[df["due_date"].notna()].copy()
df_with_due_date = df_with_due_date.sort_values("due_date")

if len(df_with_due_date) > 0:
    fig_timeline = px.timeline(
        df_with_due_date,
        x_start="due_date",
        x_end="due_date",
        y="status",
        color="priority",
        hover_data=["title", "assignee", "chat_name"],
        title="<b>Task Due Date Timeline</b>",
        color_discrete_map=priority_colors,
        labels={"priority": "Priority", "status": "Status"},
    )

    fig_timeline.update_layout(
        height=500,
        width=1200,
        xaxis_title="Due Date",
        yaxis_title="Status",
        title_font_color="#ffffff",
        xaxis_title_font_color="#ffffff",
        yaxis_title_font_color="#ffffff",
        plot_bgcolor="#1e1e1e",
        paper_bgcolor="#111111",
    )
    fig_timeline.update_xaxes(tickfont_color="#ffffff")
    fig_timeline.update_yaxes(tickfont_color="#ffffff")

    fig_timeline.show()
    print(f"\nTimeline: Showing {len(df_with_due_date)} tasks with due dates")
else:
    print("No tasks with due dates to display in timeline")
```
